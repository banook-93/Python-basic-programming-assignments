7. Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to
its next() method: 2, 3, 5, 7, 11, ...

```python
def genPrimes():
    primes = []
    num = 2
    
    while True:
        is_prime = True
        for p in primes:
            if num % p == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(num)
            yield num
        
        num += 1

primes = genPrimes()
print(next(primes))
print(next(primes))
```
